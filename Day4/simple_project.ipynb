### Import Libraires

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

### Import Data

```python
iris = datasets.load_iris()
```

```python
x = iris.data[:100 ,:2]
y = iris.target[:100] 
```

### visualization

```python
plt.figure(figsize=(8, 6))
plt.scatter(x[y==0, 0], x[y==0, 1], color='red', label='Setosa')
plt.scatter(x[y==1, 0], x[y==1, 1], color='blue', label='Versicolor')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.title('Iris Dataset (First 2 Features - 100 Samples)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame(iris.data[:100, :2], columns=['sepal length (cm)', 'sepal width (cm)'])
df['species'] = y  # y = iris.target[:100]
df['species'] = df['species'].map({0: 'Setosa', 1: 'Versicolor'})

sns.pairplot(df, hue='species')
plt.suptitle('Pairplot - First 2 Features (First 100 Samples)', y=1.02)
plt.show()
```

### Split Data

```python
x_train , x_test , y_train ,y_test = train_test_split(x , y , test_size=0.2 , random_state=42) 
```

### Train Mdoel

```python
model = LogisticRegression()
model.fit(x_train , y_train)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
y_pred
```

### Evaluation


```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

```python
accuracy1 = accuracy_score(y_test , y_pred)
accuracy1
```

```python
cm = confusion_matrix(y_test , y_pred)
print(cm)
```

```python
report = classification_report(y_test , y_pred)
print(report)
```
